### 4. Data Preparation: Sourcing, Cleaning, and Merging
The foundation of any analysis is a clean, well-structured dataset. This phase involved a multi-step process to transform several raw, disparate files into a single, analysis-ready master DataFrame.

**Process:**
* **Sourcing & Acquisition:** Data was gathered from four different sources. The process involved navigating data portals and, when direct links proved unstable, pivoting to manual downloadsâ€”a common real-world challenge.
* **Individual Cleaning:** Each of the five raw datasets was loaded into a pandas DataFrame and cleaned individually. This included filtering for our analysis period (2000-2020), selecting only relevant columns, and standardizing column names for consistency (e.g., renaming `Entity` to `Country_Name`). A key data cleaning step involved inspecting the raw malaria data and selecting the clean `No. of cases_median` column for analysis over the raw text field.
* **Merging:** The cleaned DataFrames were sequentially merged using a `left` join to ensure all health and country data was preserved. The final result is a single, rich DataFrame, `df_project2_final`.

```python
import pandas as pd

# --- Load the three datasets ---

# Replace with your actual filenames
temp_file = 'annual-temperature-anomalies.csv'
water_file = 'population-using-at-least-basic-drinking-water.csv'
gbd_file = 'cause_of_deaths.csv' # This name might be slightly different for you

try:
    # Load into pandas DataFrames
    df_temp_raw = pd.read_csv(temp_file)
    df_water_raw = pd.read_csv(water_file)
    df_gbd_raw = pd.read_csv(gbd_file)

    print("--- 1. Temperature Anomaly Data ---")
    print(f"Shape: {df_temp_raw.shape}")
    print(df_temp_raw.info())
    print(df_temp_raw.head())

    print("\n--- 2. Water Access Data ---")
    print(f"Shape: {df_water_raw.shape}")
    print(df_water_raw.info())
    print(df_water_raw.head())

    print("\n--- 3. Global Burden of Disease (GBD) Data ---")
    print(f"Shape: {df_gbd_raw.shape}")
    # For very large files, .info() can be slow, but it's essential here
    print(df_gbd_raw.info())
    print(df_gbd_raw.head())

except FileNotFoundError as e:
    print(f"Error: A file was not found. Please double-check your filenames.")
    print(e)
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# --- Clean Temperature Data ---

# Filter for the years 2000-2020
df_temp_clean = df_temp_raw[(df_temp_raw['Year'] >= 2000) & (df_temp_raw['Year'] <= 2020)].copy()

# Select and rename the columns we need
df_temp_clean = df_temp_clean[['Entity', 'Year', 'Temperature anomaly']]
df_temp_clean.rename(columns={
    'Entity': 'Country_Name',
    'Temperature anomaly': 'Temp_Anomaly_C'
}, inplace=True)

print("--- Clean Temperature DataFrame ---")
print(df_temp_clean.info())
print(df_temp_clean.head())
```

```python
# --- Clean Water Access Data ---

# Filter for the years 2000-2020
df_water_clean = df_water_raw[(df_water_raw['Year'] >= 2000) & (df_water_raw['Year'] <= 2020)].copy()

# Select and rename the columns we need
df_water_clean = df_water_clean[['Entity', 'Year', 'Share of the population using at least basic drinking water services']]
df_water_clean.rename(columns={
    'Entity': 'Country_Name',
    'Share of the population using at least basic drinking water services': 'Water_Access_Percent'
}, inplace=True)

print("\n--- Clean Water Access DataFrame ---")
print(df_water_clean.info())
print(df_water_clean.head())
```

```python
# --- Clean GBD Data ---

# Filter for the years 2000-2020
df_gbd_clean = df_gbd_raw[(df_gbd_raw['Year'] >= 2000) & (df_gbd_raw['Year'] <= 2020)].copy()

# Select and rename the columns we need
# Note: We are using the columns available in the file you downloaded.
df_gbd_clean = df_gbd_clean[[
    'Country/Territory',
    'Year',
    'Lower Respiratory Infections',
    'Nutritional Deficiencies',
    'Diarrheal Diseases'
]]

df_gbd_clean.rename(columns={
    'Country/Territory': 'Country_Name',
    'Lower Respiratory Infections': 'Deaths_LRI',
    'Nutritional Deficiencies': 'Deaths_Nutritional_Deficiencies',
    'Diarrheal Diseases': 'Deaths_Diarrheal_Diseases'
}, inplace=True)

print("\n--- Clean Global Burden of Disease DataFrame ---")
print(df_gbd_clean.info())
print(df_gbd_clean.head())
```

```python
# --- Merge the three clean DataFrames ---

# 1. Merge the temperature and water access data first
# We merge on 'Country_Name' and 'Year'
df_merged = pd.merge(df_temp_clean, df_water_clean, on=['Country_Name', 'Year'], how='outer')

# 2. Merge the result with the GBD health data
df_final = pd.merge(df_merged, df_gbd_clean, on=['Country_Name', 'Year'], how='outer')


# --- Inspect the Final Merged DataFrame ---
print("--- Final Merged DataFrame ---")
print("Merge complete.")

print("\nInfo of the final DataFrame:")
df_final.info()

print("\nHead of the final DataFrame:")
print(df_final.head())

# Let's also check for missing values in each column
print("\nMissing values count in each column:")
print(df_final.isnull().sum())
```

```python
import pandas as pd

# --- Final Corrected Step 1: Loading and Cleaning Income Group Data ---

print("Loading local income group data from the uploaded Excel file...")

try:
    # 1. Define the filename that you uploaded to Colab
    income_group_filename = 'OGHIST.xlsx'

    # 2. Load the data from the local Excel file
    df_income_raw = pd.read_excel(income_group_filename,
                                  sheet_name='Country Analytical History',
                                  skiprows=5)

    print("Successfully loaded the raw Excel file.")

    # 3. Clean and Reshape the Income Data
    id_vars = ['Unnamed: 0']
    year_vars = [col for col in df_income_raw.columns if isinstance(col, int) and 2000 <= col <= 2020]
    df_income_subset = df_income_raw[id_vars + year_vars]

    df_income_long = pd.melt(df_income_subset,
                             id_vars=id_vars,
                             var_name='Year',
                             value_name='Income_group')

    df_income_long.rename(columns={'Unnamed: 0': 'Code'}, inplace=True)

    # --- THIS IS THE CORRECTED PART ---
    # Instead of getting the map from df_final, we get it from df_temp_raw, which we know has both columns.
    print("Creating a country name to country code map...")
    country_map = df_temp_raw[['Entity', 'Code']].drop_duplicates().dropna()
    country_map.rename(columns={'Entity': 'Country_Name'}, inplace=True)
    # --- END OF CORRECTION ---

    # Now we can add the full country name to our income data
    df_income_final = pd.merge(df_income_long, country_map, on='Code', how='left')

    print("Successfully cleaned and reshaped the income data.")

    # 4. Merge this into our final DataFrame
    df_final_with_income = pd.merge(df_final, df_income_final[['Country_Name', 'Year', 'Income_group']], on=['Country_Name', 'Year'], how='left')

    print("\nSuccessfully added income group data to the final DataFrame.")
    print("Final DataFrame Info:")
    print(df_final_with_income.info())

except FileNotFoundError:
    print(f"Error: The file '{income_group_filename}' was not found.")
    print("Please make sure you have uploaded the file to Colab and the filename matches exactly.")
except Exception as e:
    print(f"\nAn unexpected error occurred: {e}")
```

With a clean dataset prepared, the analysis phase began by exploring the relationships between our variables.

#### Initial Approach: A Global Correlation Analysis
The initial hypothesis was that a global-level analysis would show a direct correlation between climate factors and health outcomes. To test this, a correlation heatmap was generated.

**Finding 1: Global Correlations are Weak and Misleading.**
The heatmap revealed a critical insight: the correlations between our primary climate/resilience variables and the health outcomes were surprisingly weak (mostly between -0.2 and 0.1). This indicated that a simple global analysis was "washing out" the real trends and that the relationships must be highly context-dependent.

#### Pivoting the Approach: The Case Study Narrative
Discovering the weakness of the global analysis was not a failure, but a finding in itself. It prompted a pivot in strategy away from broad, noisy scatter plots toward a more focused **"Case Study" approach**.

**Rationale:** The most powerful stories are often found by examining the specific journeys of individual countries over time. This approach allows us to visually correlate spikes in climate stressors with health outcomes within a single, consistent context. To strengthen this analysis, we added a key vector-borne disease metric: **Malaria Incidence**.

#### The Final Visualizations: Country "Dashboard" Charts
The final output of this project is a series of dashboard-style time-series charts for six countries that are on the front lines of climate change. Each chart plots multiple variables, using a shared x-axis (Year) and dual y-axes to handle different scales, telling a rich story for that specific nation.

**Finding 2: The Story of Vulnerability and Resilience.**
The case study charts for nations like **Mali, Bangladesh, and Somalia** clearly visualize the story of high vulnerability.
```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# --- Final Code Block for Project 2 Analysis ---

try:
    # --- Part 1: Clean and Prepare the Malaria Data ---
    print("Step 1: Cleaning the malaria data...")

    # We already have df_malaria_test loaded in memory from the last step
    # We select and rename the columns we need, using the clean 'median' column
    df_malaria_clean = df_malaria_test[['Country', 'Year', 'No. of cases_median']].copy()
    df_malaria_clean.rename(columns={
        'Country': 'Country_Name',
        'No. of cases_median': 'Malaria_Cases_Estimated'
    }, inplace=True)

    print("Malaria data cleaned successfully.")

    # --- Part 2: Final Merge ---
    print("\nStep 2: Merging malaria data into the main DataFrame...")

    # Merge the clean malaria data with our main DataFrame from the previous steps
    df_project2_final = pd.merge(df_final_with_income, df_malaria_clean, on=['Country_Name', 'Year'], how='left')

    print("Final merge complete. Dataset is ready for visualization.")


    # --- Part 3: Create a Reusable Plotting Function ---
    print("\nStep 3: Defining the case study plotting function...")

    def plot_country_case_study(country_name):
        """
        Generates a dual-axis time-series chart for a selected country.
        """
        # Select data for the chosen country
        df_country = df_project2_final[df_project2_final['Country_Name'] == country_name].copy()

        # Filter for the best year range with the most overlapping data (2000-2017 is often good for these sources)
        df_country = df_country[(df_country['Year'] >= 2000) & (df_country['Year'] <= 2017)]

        if df_country.empty:
            print(f"No sufficient data to plot for {country_name}.")
            return

        # Setup the plot
        fig, ax1 = plt.subplots(figsize=(14, 8))
        plt.title(f'Climate and Health Trends in {country_name} (2000-2017)', fontsize=16)

        # Plot 1: Temperature Anomaly on the first y-axis (ax1)
        color = 'tab:red'
        ax1.set_xlabel('Year')
        ax1.set_ylabel('Temperature Anomaly (Â°C)', color=color, fontsize=12)
        ax1.plot(df_country['Year'], df_country['Temp_Anomaly_C'], color=color, marker='o', linestyle='--', label='Temperature Anomaly')
        ax1.tick_params(axis='y', labelcolor=color)
        ax1.grid(False)

        # Create a second y-axis that shares the same x-axis
        ax2 = ax1.twinx()

        # Plot 2: Malaria Cases on the second y-axis (ax2)
        color = 'tab:blue'
        ax2.set_ylabel('Disease Cases / Deaths (Absolute Numbers)', fontsize=12)
        # We plot Malaria Cases first
        ax2.plot(df_country['Year'], df_country['Malaria_Cases_Estimated'], color=color, marker='s', label='Malaria Cases (Estimated)')
        ax2.tick_params(axis='y')

        # Plot 3: Diarrheal Deaths on the second y-axis (ax2)
        color = 'tab:green'
        # We plot Diarrheal Deaths on the same axis
        ax2.plot(df_country['Year'], df_country['Deaths_Diarrheal_Diseases'], color=color, marker='^', label='Diarrheal Deaths')

        # Add legends from both axes
        fig.legend(loc="upper right", bbox_to_anchor=(0.9, 0.9))

        plt.show()

    print("Plotting function defined.")

    # --- Part 4: Generate the Case Study Visualizations ---
    print("\nStep 4: Generating charts for our case study countries...")

    plot_country_case_study('Mali')
    plot_country_case_study('Colombia')
    plot_country_case_study('Indonesia')

except Exception as e:
    print(f"An error occurred: {e}")
```
